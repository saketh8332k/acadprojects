```python
#data cleaning-(4)removing outliers
def remove(dataframe,column):
    Q3 = dataframe[column].quantile(0.75)
    Q1 = dataframe[column].quantile(0.25)
    IQR = Q3 - Q1
    upper = Q3 + (1.5 * IQR)
    lower = Q1 + (1.5 * IQR)
    df_no_outliers = dataframe[(dataframe[column]>lower) & (dataframe[column]<upper)]
    return df_no_outliers
```

```python
#pd.get_dummies()-converts categorical data into dummy or indicator variables(that r binary(0,1)).used in ml when machine needs numerical i/p.
final_data=pd.get_dummies(train_data,drop_first=True)
```

```python
#preparing data for model.x = independent variables,y = dependent variable.
X=final_data.drop(columns=['Item_Outlet_Sales'],axis=1)
Y=final_data['Item_Outlet_Sales']
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error,accuracy_score
```

```python
xgb_reg = XGBRegressor()
xgb_reg.fit(X_train,Y_train)
score = xgb_reg.score(X_train,Y_train)
print(f"Score: {score}")
```

```python
# Calculate evaluation metrics correctly
r2 = r2_score(Y_train, xgb_reg.predict(X_train))  # Pass actual and predicted target values
mse = mean_squared_error(Y_train, xgb_reg.predict(X_train))/100000  
rmse = np.sqrt(mse)  
mae = mean_absolute_error(Y_train, xgb_reg.predict(X_train))

# Calculate Accuracy
accuracy = (1 - (mae / np.mean(Y_train))) * 100  

# Print Results
print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse: .2f}")
print(f"Root Mean Squared Error (RMSE): {rmse: .2f}")
print(f"Mean Absolute Error (MAE): {mae: .4f}")
print(f"Model Accuracy: {accuracy:.2f}%")
```
